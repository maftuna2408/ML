### Heart Disease Prediction

Step 1. Import Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Step 2. Data collection and Processing

```python
# loading the csv data to a Pandas DataFrame

heart_data = pd.read_csv(r"C:\Users\Maftuna\Downloads\heart.csv")
```

```python
# Setting full column names
heart_data.columns = [
    "Age",                         # Age of the person
    "Sex",                         # Gender
    "Chest Pain Type",             # Type of chest pain
    "Resting Blood Pressure",      # Blood pressure at rest
    "Serum Cholesterol",           # Cholesterol level in the blood
    "Fasting Blood Sugar",         # Blood sugar level when fasting
    "Resting Electrocardiographic Results",  # Resting ECG results
    "Maximum Heart Rate Achieved", # Maximum heart rate achieved
    "Exercise-Induced Angina",     # Angina induced by exercise
    "ST Depression Induced by Exercise",     # ST depression caused by exercise
    "Slope of the Peak Exercise ST Segment", # Slope of the ST segment
    "Number of Major Vessels Colored by Fluoroscopy", # Major vessels colored during fluoroscopy
    "Thalassemia",                 # Blood disorder (thalassemia)
    "Target (Presence of Heart Disease)" # Target variable (Heart disease presence)
]

# Verifying the updated column names
print(heart_data.columns)

```

Step 3. Split the Features and Target 

```python
X = heart_data.drop(columns='Target (Presence of Heart Disease)', axis=1)
Y = heart_data['Target (Presence of Heart Disease)']
```

```python
print(X)
```

```python
print(Y)
```

Step 4. Splitting the Data into Trainging data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print("X shape: ", X.shape)
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler




# Scalingni yaratish
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Train uchun fit va transform
X_test_scaled = scaler.transform(X_test)       # Test uchun faqat transform
```

Step 5. Model Training

```python
# Logictic Regression 

model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score 

```python
# accurancy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training data : " , training_data_accuracy)
```

```python
# accurancy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Test data : " , test_data_accuracy)
```

Step 6. Building a Predictive System

```python
input_data = (51,0,0,130,305,0,1,142,1,1.2,1,0,3)

# Change the input data to a array 
input_data_as_numpy_array = np.array(input_data)

# Reshape the numpy array as we are predicting for only on instance 
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction[0] == 0:
    print("The Person does not have a Heart Disease")
else: 
    print("The Person has Heart Disease")
```

```python
# Logistic Regression Coefficients
coefficients = model.coef_
print("Model Coefficients:", coefficients)

# Feature Importance for Decision Trees or Random Forests
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train_scaled, Y_train)
importances = rf_model.feature_importances_
print("Feature Importances:", importances)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# k-Nearest Neighbors modeli
knn_model = KNeighborsClassifier()
knn_model.fit(X_train_scaled, Y_train)
y_pred_knn = knn_model.predict(X_test_scaled)

accuracy_knn = accuracy_score(Y_test, y_pred_knn)
print("Accuracy with KNN (scaled):", accuracy_knn)
```
