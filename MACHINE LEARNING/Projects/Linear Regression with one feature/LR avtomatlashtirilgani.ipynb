### Yuqoridagi codelarni hammasini avtomatlashtirish

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Maftuna\Desktop\Machine Learning\Projects\Linear Regression with one feature\ml.csv")
df.head()
```

```python
df.rename(columns={'Unnamed: 1':'y'}, inplace=True)
df.head()
```

```python
x = np.array(df['x'])
y = np.array(df['y'])
m = len(y) 
print("Namunalar soni: ", m)
```

```python
iteration = 100 # nechi marta qayta takrorlanish jarayoni (100 marta iteration)
alpha = 0.0001 # learning rate
theta_0 = 0
theta_1 = 0
Loss_list = []

for i in range(iteration):
    
    ## Hypothesis function (Gipotezani hisoblash)
    h_x = theta_0 + theta_1 * x

    ## Loss function (Xatolikni hisoblash)
    Loss = np.sum((h_x - y)**2) / (2*m)
    Loss_list.append(Loss)

    ## Gradient descent (Gradientni hisoblash)
    theta_0 = theta_0 - alpha * np.sum(h_x - y) / m
    theta_1 = theta_1 - alpha * np.sum((h_x - y) * x) / m

    ## print loss
    if i % 10 == 0:
        print(f"Loss qiymati {i} qadamda:  {Loss}")

print(f"\nLoss: {Loss}")
print(f"Theta_0: {theta_0}")
print(f"Theta_1: {theta_1}")
```

```python
y_loss = np.array(Loss_list)
x_axis = range(iteration)

plt.plot(x_axis, y_loss)
plt.xlabel('Qadamlar (Iteration)')
plt.ylabel('Loss qiymati')
plt.title('Loss function')
plt.show()
```

#### Theta_0: 0.109  bu y-intercept qiymati.
#### Theta_1: 2.134  bu gradient (slope) qiymati.

```python
# alphani o'zgartirish
# alpha = 0.00001
# iteration = 500 

iteration = 500 # nechi marta qayta takrorlanish jarayoni (100 marta iteration)
alpha = 0.00001 # learning rate
theta_0 = 0
theta_1 = 0
Loss_list = []

for i in range(iteration):
    
    ## Hypothesis function (Gipotezani hisoblash)
    h_x = theta_0 + theta_1 * x

    ## Loss function (Xatolikni hisoblash)
    Loss = np.sum((h_x - y)**2) / (2*m)
    Loss_list.append(Loss)

    ## Gradient descent (Gradientni hisoblash)
    theta_0 = theta_0 - alpha * np.sum(h_x - y) / m
    theta_1 = theta_1 - alpha * np.sum((h_x - y) * x) / m

    ## print loss
    if i % 10 == 0:
        print(f"Loss qiymati {i} qadamda:  {Loss}")

print(f"\nLoss: {Loss}")
print(f"Theta_0: {theta_0}")
print(f"Theta_1: {theta_1}")
```

```python
#   Loss qiymatining 139 dan keyin deyarli ozgarmasligi


#   Model optimal yoki local minimumga yetgan.
#   Oqitishni toxtatishingiz mumkin, chunki natijani yaxshilash qiyin.
#   Learning rate yaxshi tanlangan:

#   Alpha juda katta yoki kichik emas, shuning uchun loss sekin-asta tushgan va stabil bolgan.
#   Iteratsiyalarni kopaytirish kerak emas:

#   139 dan keyin deyarli ozgarmagan loss qiymatlari shuni korsatadiki, iteratsiyalarni kopaytirishdan foyda yoq.
```

```python
y_loss = np.array(Loss_list)
x_axis = range(iteration)

plt.plot(x_axis, y_loss)
plt.xlabel('Qadamlar (Iteration)')
plt.ylabel('Loss qiymati')
plt.title('Loss function')
plt.show()
```

Kichik alpha: Har bir iteratsiyada kichikroq qadam tashlanadi.
Alpha ning tasiri
1. Juda kichik alpha
Tasiri:
Model sekin konvergensiya (natijaga yetish) qiladi.
Iteratsiyalar soni oshadi, chunki qadamlar juda kichik.
Vaqtni koproq talab qiladi, lekin maqsadga erishadi.
Muammo:
Juda kichik alphada iteratsiya kop vaqt oladi va noqulay bolishi mumkin.
2. Juda katta alpha
Tasiri:
Model optimal qiymatni o'tkazib yuboradi yoki osillatsiya qiladi.
Qadamlar juda katta bolgani uchun, optimal nuqtaga yaqinlashish orniga u orqali otib ketadi yoki tebranadi.
Muammo:
Model konverge bolmaydi va yaxshi natija bermaydi.
3. Optimal alpha
Tasiri:

Iteratsiyalar soni kam boladi va vaqtni tejaydi.
